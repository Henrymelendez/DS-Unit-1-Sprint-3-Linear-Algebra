## 1.1 Create a two-dimensional vector and plot it on a graph

```
v_1 = np.array([.7,.5])
v_2= np.array([.6,.7])

fig, ax = plt.subplots()
ax.grid()
plt.xlim(-.01,.9)
plt.ylim(-.01,.9)

for color, vector in [('red',v_1),('green',v_2)]:
  plt.arrow(0,0,vector[0],vector[1], head_width=.02, head_length=.01, linewidth = 2, color=color)

plt.show()
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.mat('12 11 ; 7 10')
J = np.mat('0 1 2 ; 7 10 4 ; 3 2 0')
print('H:',np.linalg.det(H))
print('J:',np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
inv_h = np.linalg.inv(H)
pd.DataFrame(inv_h)
```

```
inv_j = np.linalg.inv(J)
pd.DataFrame(inv_j)
```
